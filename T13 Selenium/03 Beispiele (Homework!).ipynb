# Bsp2 Benzinpreise

```python

driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')
url = "http://www.benzin-preis.ch/"

driver.get(url)
time.sleep(2)
search = driver.find_element_by_name('search')
search.click()

count = 6
for element in range(20):
    button = driver.find_element_by_xpath("//a[@href='?next="+ str(count) +"']").click()
    page = driver.page_source.encode('utf-8')
    with open('benzin_' + str(element) + '.htm', 'wb+') as file:
        file.write(page)
        file.close()
    button
    count += 5
    time.sleep(2)

    
# Status: Alle Files abgespeichert. Done!! =)
```

```python
leere_liste = []

for entry in day:
    today = date.today()
    day_date = entry.find('div', {'class':'day-date'}).text
    day_price = entry.find('div', {'class':'day-price'}).text
    day_week = entry.find('div', {'class':'day-week'}).text

    
    mini_dict = {
        "Zugriffsdatum": date.today(),
        "Datum": day_date,
        "Preis": day_price,
        "Wochentag": day_week}
    leere_liste.append(mini_dict)

    
df = pd.DataFrame(leere_liste)
df
```
