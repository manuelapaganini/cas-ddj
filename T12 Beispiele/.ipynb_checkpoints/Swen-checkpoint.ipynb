# Arzneilisten

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
url = requests.get('https://www.iqwig.de/de/presse/amnog-uebersicht.7723.html')
soup = BeautifulSoup(url.content,'lxml')

trlist = soup.find_all("tr")

medis = []

for element in trlist[1:]:
    td_list = element.find_all('td')
    datum = td_list[0].text
    wirkstoff = td_list[1].text
    anwendungsgebiet = td_list[2].text
    ergebnis = td_list[3].text
    anmerkung = td_list[4].text
    
    mini_dict = {"Datum": datum,
        "Wirkstoff": wirkstoff,
        "Anwendungsgebiet": anwendungsgebiet,
        "Ergebnis": ergebnis,
        "Anmerkung":anmerkung}
    
    medis.append(mini_dict)

df = pd.DataFrame(medis)
df.head()
```
