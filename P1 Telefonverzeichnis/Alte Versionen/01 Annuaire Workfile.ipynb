```python

file = open('Sourcefiles/Telefonverzeichnis 1.html', 'r')
text = file.read()


def Verzeichnis (text_):
    

    soup = BeautifulSoup(text_, 'lxml')
    
    td_liste = soup.find_all('td', {'role':"gridcell"})
    
    lst = []
    
    for gridcell in td_liste:
        
        lst.append(gridcell.text)
    
    return lst
file.close()
```
