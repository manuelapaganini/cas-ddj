# Worksheet File Reading

Eine wunderschöne Datei öffnen und lesen:

```python
file = open("lesen.txt", "r")
for line in file:
    print(line.strip())
file.close()
```

In eine megatolle Datei schreiben:

```python
file = open("schreiben.txt", "w")

students = ["alpha", "beta", "gamma"]

for student in students:
    file.write(student + "\n")

file.close()
```

Eine Datei mit with öffnen

```python
with open("schreiben.txt", "r") as file:
    for line in file:
        print(line.strip())
```

Eine csv-Datei öffnen:

```python
with open("datei.csv", "r") as file:
    for line in file:
        data = line.split(";")
        print(data[0] + ": " + data[1])
```

Bedingungen fürs File-Lesen:

```python
with open("datei.csv", "r") as file:
    for line in file:
        data = line.split(";")
        if data[0] == "Muenchen":
            print(data[0], data[1])
```

Zwischenschritte für Hunderechnung:

```python
jahr = []

with open("20151001_hundenamen.csv", "r") as file:
    for line in file:
        data = line.split(",")
        jahr.append(data[1])
        # print(data[0] + ": " + data[1])
    print(jahr)
    
```
