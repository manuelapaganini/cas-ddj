```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://daslamm.ch/"

r = requests.get(url)
soup = BeautifulSoup(r.content, 'lxml')
titles = soup.find_all('h2')

lst = []
for title in titles:
    lst.append(title.text)
```

```python
df = pd.DataFrame(lst)
df
```

```python
df.to_csv("das_lamm_titel.csv")
```
