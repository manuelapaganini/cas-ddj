### Task 1: stopwords aus einem Trainingsdatensatz filtern

```python
# Import stuff

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
from io import open

stopwords = stopwords.words('german')

import os,glob
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

from textblob.classifiers import NaiveBayesClassifier
```

    Wir machen eine Trainingsliste. In der Country-Liste sind die Ordnernamen.

    wir gehen durch Files in den beiden Ordnern.  Die Ordner öffnen wir mit('songtexte/[Ordnername]')
    mit glob machen gucken wir in die Dateien
    mit open öffnen wir die Dateien
    Wir tokenizen die Songtexte, wandeln sie in Lowercase um
    fügen das Resultat der Out-Liste an.
    
    Dann streichen wir mit set() Doppel aus den Stichworten. 
    Dann packen wir die out-Elemente in die Train-Liste. Und weil wir immer noch ind er Ordner-Loop sind, kann
        er dazu das Land packen. 




```python
train = []
countries = ["schweiz", "deutschland"]
for country in countries:
    out = []
    folder_path = 'songtexte/%s' % country 
    for filename in glob.glob(os.path.join(folder_path, '*.txt')):
      with open(filename, 'r') as f:
        text = f.read()
        words = word_tokenize(text)
        words=[word.lower() for word in words if word.isalpha()]
        for word in words:
            out.append(word)
    out = set(out)
    for word in out:
        train.append((word, country))
        #print (filename)
        #print (len(text))
train
```

```python
training_new = []

for element in train:
    words = word_tokenize(element[0])
    words = [word.lower() for word in words if word.isalpha()]
    words = [word for word in words if word not in stopwords]
    training_new.append((" ".join(words), element[1]))
training_new
```

```python

c2 = NaiveBayesClassifier(training_new)
```

```python
c2.classify("Ich gehe durch den Wald")  # "deutsch"
```

```python
c2.classify("können wir in die Mittagspause gehen")  # "schweiz"
```

```python
c2.show_informative_features()
```
