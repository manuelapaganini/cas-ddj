# Komplizierte JSONs auslesen

```python
import pandas as pd
import numpy as np
import glob
import json
%matplotlib inline
```

Jsons sind Dictionaries mit Listen mit weiteren Dictionaries

```python
path = 'psyreg_output/'
filenames = glob.glob(path + '*.json')
```

```python
len(filenames)
```

```python
data = []
with open(filenames[0]) as json_data:
    data = json.load(json_data)
data
```

Eine der verschachtelten Listen von Dicts ist zum Beispiel das Feld 'bewilligungErteilt'. 

```python
type(data["bewilligungenErteilt"]) == list
```

```python
# Wir lassen mal alle key & value-pairs ausprinten.

for k, v in data.items():
    print("%s: %s" %(k,v))
```

```python
# Schwierig daran ist, dass ein DF damit nur begrenzt Sinn macht -> Weil die verschachtelten Listen 
# als [] angezeigt werden

liste = []
minidict = {}
for k, v in data.items():
    minidict[k] = v
liste.append(minidict)
pd.DataFrame.from_dict(liste)
```
