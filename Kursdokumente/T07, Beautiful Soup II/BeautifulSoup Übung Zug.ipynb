```python
# Wir bauen ein mini-dictionary für die verschiedenen Zellen. Dafür legen wir Variablen an
# mit denen steuern wir die einzelnen Element der Zeilen-Liste an. 

leere_liste = []



for line in tr_list:
    td_list = line.find_all('td')
    href = td_list[0].find('a')['href']
    text = td_list[1].text
    stand = td_list[2].text
    zuständig = td_list[3].text
    geschäft = td_list[4].text
    status = td_list[5].text

    mini_dict ={"Link": href,
                "text": text,
                "stand": stand,
                "Zuständig": zuständig,
                "Geschäft": geschäft,
                "Status":status}
    leere_liste.append(mini_dict)
    
pd.DataFrame(leere_liste)
```
