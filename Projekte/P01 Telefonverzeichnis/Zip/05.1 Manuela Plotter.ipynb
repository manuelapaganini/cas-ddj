# Daten Plotten

Jetzt geht es daran, die Daten sichtbar zu machen.


```python
# import stuff

import pandas as pd
%matplotlib inline
```

```python
# Die Ursprungsdatei war Verzeichnis_brut. Darin habe ich mit Numbers doppelte Einträge zusammengefügt und
# die Anrede umgewandelt. 

df = pd.read_csv('Verzeichnis_per_neu.csv')
```

```python
df.head()
```

```python
df.groupby('Direktion')['Gender'].value_counts()
```

```python
# Alle Mitarbeitende nach Direktion & Gender

df.groupby('Direktion')['Gender'].value_counts().unstack().plot(kind='bar', figsize=(12,6))
```

## Einteilung nach Kader

Jetzt geht es darum, das Kader einzuteilen. Dafür wechsle ich auf Excel.


```python
# Das bereinigte CSV importieren:

df_k = pd.read_csv('Verzeichnis_kad.csv')
```

```python
# Kader nach Direktion & Gender

df_k[df_k['Kader'] == 'TRUE'].groupby('Direktion')['Gender'].value_counts()
```

```python

df_k[df_k['Kader'] == 'TRUE'].groupby('Direktion')['Gender'].value_counts().unstack().plot(kind='bar', figsize=(12,6))
```
