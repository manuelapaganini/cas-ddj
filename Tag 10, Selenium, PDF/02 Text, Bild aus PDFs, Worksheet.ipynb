## Alles zusammen

```python
lst = []

for pdf in tqdm(pdfs):
    try: 
        text = textract.process('pdfs/' + pdf, method='pdfminer', encoding='ISO-8859-1') #method='pdfminer' encoding='ISO-8859'
        text = str(text).replace('\n', ' ')
        if len(text) <40:
            text = textract.process('pdfs/'+pdf, method= 'tesseract', encoding='ISO-8859-1')
            text = text.decode('ISO-8859-1').replace('\n', ' ')
            
        mini_dict = {'Text': text,
                    'File': pdf}
        lst.append(mini_dict)
    except:
        mini_dict = {'Text': 'Fehlermeldung',
                    'File': pdf}        
        lst.append(mini_dict)
        
```

```python
df = pd.DataFrame(lst)
```

```python
df.to_csv('Export_1.csv')
```

```python
def length(elem):
    elem= len(elem)
    return elem

df['Text Länge'] = df['Text'].apply(length)
```
